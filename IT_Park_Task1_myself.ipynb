```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Educational centers'], df['Percentage Fluent Students'])
plt.xlabel('Number of Educational Centers')
plt.ylabel('Percentage of Fluent Students')
plt.title('Relationship between Educational Centers and Percentage of Fluent Students')
plt.grid(True)

# Add the regression line
# Perform linear regression
X = df['Educational centers']
y = df['Percentage Fluent Students']
X = sm.add_constant(X) # Add a constant term for the intercept

model = sm.OLS(y, X).fit()

# Get the model parameters
slope = model.params['Educational centers']
intercept = model.params['const']

# Generate y values based on the regression equation
x_values = np.linspace(df['Educational centers'].min(), df['Educational centers'].max(), 100)
y_values = intercept + slope * x_values
plt.plot(x_values, y_values, color='red', label=f'Regression Line (y = {intercept:.2f} + {slope:.4f}x)')

plt.legend()
plt.show()

correlation = df['Educational centers'].corr(df['Percentage Fluent Students'])
print(f"The correlation coefficient between Educational Centers and Percentage of Fluent Students is: {correlation:.2f}")
```

The Scatterplot and Linear Regression Model above show that the number of Educational centers in regions is directly proportional to the percentage of fluent in foreign languages students in every region. It proposes the obvious idea that **by increasing the number of Language Educational centers we can achieve higher number of students who can speak several languages**, especially the ones needed for the IT Park's initiative.

```python
regions = df['Region']
results = {
    'Residents share': percentage_residents.tolist(),
    'Fluent students share': percentage_fluent_students.tolist(),
    'Educational centers share': percentage_centers.tolist()
}


def shares_percentage(results, regions):
    labels = list(results.keys())
    data = np.array(list(results.values()))
    data_cum = data.cumsum(axis=1)
    category_colors = plt.colormaps['RdYlGn'](np.linspace(0.15, 0.85, data.shape[1]))

    fig, ax = plt.subplots(figsize=(21, 5)) # Increased width further
    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, np.sum(data, axis=1).max())

    for i, (colname, color) in enumerate(zip(regions, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        rects = ax.barh(labels, widths, left=starts, height=0.5, label=colname, color=color)

        r, g, b, _ = color
        text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
        #ax.bar_label(rects, label_type='center', color=text_color)
    ax.legend(ncols=len(regions), bbox_to_anchor=(0, 1), loc='lower left', fontsize='small')

    return fig, ax


shares_percentage(results, regions)
plt.show()
```

The colored graph visualized the share for Residents, Fluent Students and Educational centers across regions. Also, it makes a comparison in between these 3 categories. It proposes an idea that despite Tashkent having more fluent students and educational centers, **it still needs improvements**, as the proportion of Residents who need these kind of people is even higher. Another proposal is **increasing the number of Residents in regions**. This way we can achieve the balance in demand and offer.

# Resident companies in the future

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Prepare data for polynomial regression
X_train = df_residents[['Years']]
y_train = df_residents['Residents']

# Use polynomial features (degree 2 or 3 seems appropriate given the curve)
poly = PolynomialFeatures(degree=3)
X_train_poly = poly.fit_transform(X_train)

# Create and train the polynomial regression model
model_poly = LinearRegression()
model_poly.fit(X_train_poly, y_train)

# Predict for 2026 and 2027
years_to_predict = np.array([[2026], [2027]])
years_to_predict_poly = poly.transform(years_to_predict)
predicted_residents_poly = model_poly.predict(years_to_predict_poly)

# Add predictions to the DataFrame for plotting
df_residents_predicted_poly = df_residents.copy()
new_rows_poly = pd.DataFrame({'Years': [2026, 2027], 'Residents': predicted_residents_poly})
df_residents_predicted_poly = pd.concat([df_residents_predicted_poly, new_rows_poly], ignore_index=True)

# Calculate percentage increase from 2025
residents_2025 = df_residents[df_residents['Years'] == 2025]['Residents'].iloc[0]
percentage_increase_2026_poly = ((predicted_residents_poly[0] - residents_2025) / residents_2025) * 100
percentage_increase_2027_poly = ((predicted_residents_poly[1] - residents_2025) / residents_2025) * 100

print(f"Predicted residents in 2026 (Polynomial Regression): {predicted_residents_poly[0]:.0f}")
print(f"Predicted residents in 2027 (Polynomial Regression): {predicted_residents_poly[1]:.0f}")
print(f"Percentage increase from 2025 to 2026 (Polynomial Regression): {percentage_increase_2026_poly:.2f}%")
print(f"Percentage increase from 2025 to 2027 (Polynomial Regression): {percentage_increase_2027_poly:.2f}%")

# Plot the original data and the polynomial regression predictions
plt.figure(figsize=(10, 6))
plt.plot(df_residents_predicted_poly['Years'], df_residents_predicted_poly['Residents'], marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Number of Residents')
plt.title('Change in Residents Numbers Over the Years with Polynomial Regression Predictions')
plt.grid(True)
plt.show()
```

Here using Polynomial Regression we can predict the number of residents in the coming years and percentage increase. We will use this for further analysis. We see that expected **increase 31.77% in the number of residents for the next year**. The data above was taken from IT Park's official website.

As people who will work in IT Park's resident companies should be people who are fluent in at least one of the foreign languages above, we can assume that increase in their demand should be equivalent to increase in demand for fluent in foreign languages people.
So, we simply can assume with the data we have (as we lack more complex relations and further detailed statistics) that we will observe 31.77% increase in need for people who study foreign languages too.

We consider this increase as something natural that will happen over time, as the demand increases. We will see other relations that may affect it too and keep this 31.77% increase for the next year in our mind to further add it in our calculations.

# Balancing the different languages share according to the needs

From our "Residents Share" Pie Chart and "Share of languages spoken by students" Bar Chart, we see that the share of residents and share of fluent students in the given particular languages are not fair. As almost half of our residents require skills in those languages too, while the share of fluent students is comparatively low.
So, now we will calculate if we have enough people/students in those other foreign languages (other than English) to at least cover the demand of IT Park's residents.

```python
df_residents_languages
```

From the data we have about the language share across residents and estimations we have from IT Park (f.ex. we have more German speaking residents than French, though they are in the same category; similarly we need Korean > Chinese > Japanese comparatively).
